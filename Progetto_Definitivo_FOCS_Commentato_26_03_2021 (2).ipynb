Creazione di una funzione di normalizzazione dei nomi delle università prendendo come punto di riferimento i nome delle università contenute in School and country, nel primo **if** prendiamo i valori uguali tra i 2 dataset, in caso di non compatibilità si verificano 2 casi. Nel primo cerchiamo i nome delle università del dataset da normalizzare contenuti nel dataset school and country e nel secondo l'opposto.

```
def name_normalized (column_name):
    
    for university in school_and_country['uni_name_norm']:
      if column_name == university:
        return university
      else:
        pattern = re.compile(column_name)
        occurrence = pattern.search(university)
        if occurrence: 
          return university
        else: 
          pattern2 = re.compile(university)
          occurence2 = pattern2.search(column_name)
          if occurence2:
            return university
```

## Definiamo la funzione, che ritorna il valore medio della classe nel caso in sia un range altrimenti toglie l'uguale laddove sia presente oppure ritorna il valore senza le precedenti condizioni. Tutti i valori vengono convertiti in INT

```
def rank_normalized(value):
  if '-' in value:
    pattern = re.compile("(\d+)\-(\d+)")
    pat = pattern.search(value)
    rank = (int(pat.group(1))+int(pat.group(2)))/2
    return int(rank) 
  elif '=' in value:
    rank2=value.replace('=','')
    return int(rank2) 
  else:
    return int(value)
```

```
ranking['times_ranking'] = ranking['times_ranking'].apply(rank_normalized)
ranking['shanghai_ranking'] = ranking['shanghai_ranking'].apply(rank_normalized)
ranking.head(3)
```

Definiamo la funzione possible match, con 3 condizioni:
 
*   il raking di shanghai e time contengono entrambi il range e confrontiamo i limiti superiori e inferiori tra i due range rispetto a world ranking.
*   solo per shanghai faccio il confronto con world ranking.
*   solo per times faccio il confronto con world ranking
*   Infine se i valori in shanghai times e world coincidono allora esiste il match







```
def possible_match(db):
    
    #valori
  shanghai = db['shanghai_ranking']
  times = db['times_ranking']
  world = db['world_ranking']

  if ('-' in shanghai) and ('-' in times):
    pattern_1 = re.compile("(\d+)\-(\d+)")
    occurrence_1 = pattern_1.search(times)
    lower_1 = int(occurrence_1.group(1))
    upper_1 = int(occurrence_1.group(2))
                    
    pattern_2 = re.compile("(\d+)\-(\d+)")
    occurrence_2 = pattern_2.search(shanghai)
    lower_2 = int(occurrence_2.group(1))
    upper_2 =int(occurrence_2.group(2))
                
    if world >= lower_1 and world <= upper_1 and world >= lower_2 and world <= upper_2 :
      return 'possible'
    else:
      return 'not possible'

  elif  '-' in shanghai:
    pattern_3 = re.compile("(\d+)\-(\d+)")
    occurrence_3 = pattern_3.search(shanghai)
    lower_3 = int(occurrence_3.group(1))
    upper_3 = int(occurrence_3.group(2))
    if world >= lower_3 and world <= upper_3 and str(world) == times:
      return 'possible'
    else:
      return 'not possible'

  elif  '-' in times:
    pattern_4 = re.compile("(\d+)\-(\d+)")
    occurrence_4 = pattern_4.search(times)
    lower_4 = int(occurrence_4.group(1))
    upper_4 = int(occurrence_4.group(2))
    if world >= lower_4 and world <= upper_4 and str(world) == shanghai:
      return 'possible'
    else:
      return 'not possible'
  
  elif shanghai == times and shanghai == str(world) :
    return 'match'
  else:
    return 'no match'
```

```
same_position['possible_match']=same_position.apply(possible_match, axis=1)
```

```
final = same_position[(same_position['possible_match']=='possible') | (same_position['possible_match']== 'match')]	
final
```
